```python
import numpy as np
import pandas as pd
from scipy.stats import norm
from numpy.random import randn
from numpy import random as rn
import scipy.stats as si
from matplotlib import pyplot as plt
from IPython.display import Image
%matplotlib inline
```

```python
def rand():
    d = rn.uniform(0, 1, 1)[0]
    return (d)

rand()
```

```python
def normsdist(z):
    z = si.norm.cdf(z,0.0,1.0)
    return (z)

normsdist(1.645)
```

```python
def normsinv(x):
    x = si.norm.ppf(x)
    return (x)

normsinv(0.9500)
```

```python
def normsrand():
    y = rn.randn(1)[0]
    return (y)

normsrand()
```
